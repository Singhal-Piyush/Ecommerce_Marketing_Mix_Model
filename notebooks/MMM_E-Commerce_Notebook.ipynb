```python
# function to know about number of unqiue values and null values 
def feature_summary(dataset):
    column_summary_df = pd.DataFrame(columns =['Name','Values',
                                            'Null_Values','%_Null_Value',
                                            'Nunique_Value','Data_Type'])
    l = len(dataset)
    for i in dataset.columns:
                                            
        column_summary_df.loc[len(column_summary_df),:] = [i,dataset[i].notnull().sum(),
                                                     dataset[i].isnull().sum(),
                                                     round(dataset[i].isnull().sum()/l,2),
                                                     dataset[i].nunique(),
                                                     dataset[i].dtype
                                                    ]
    return column_summary_df
```

```python
# print the feature name and summary of values present in the dataset
feature_summary(consumer_dataset)
```
